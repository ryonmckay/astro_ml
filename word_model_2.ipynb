```python
def assign_meaning(row):
    # Create an empty dictionary to store the meanings
    meanings = {}
    
    # Loop through each column that contains a planet sign
    for planet in planets.keys():
        # Get the name of the planet sign and its corresponding house
        sign_col = f"{planet.lower()}_sign"
        house_col = f"{planet.lower()}_house"
        sign = row[sign_col]
        house = row[house_col]
        
        # Get the meanings of the planet sign and its corresponding house
        sign_meanings = planets[planet][planets[planet].index(sign) + 1:][0]
        house_meanings = houses[list(houses.keys())[house-1]][house-1]
        
        # Add the meanings to the dictionary
        meanings[f"{planet} Sign"] = sign_meanings
        meanings[f"{planet} House"] = house_meanings
    
    # Loop through each column that contains a house sign
    for i in range(1, 13):
        house_col = f"house_{i}_sign"
        house = row[house_col]
        
        # Get the meanings of the house sign
        house_meanings = houses[list(houses.keys())[i-1]][i-1]
        
        # Add the meanings to the dictionary
        meanings[f"House {i} Sign"] = house_meanings
        
    return meanings
```
